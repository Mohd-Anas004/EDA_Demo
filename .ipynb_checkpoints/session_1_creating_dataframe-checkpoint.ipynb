**inplace**

* when we do the operation the result output is not overwriting
* means the original data frame is same as it is
* in order to do this we need to use inplace argument
* by default inplace = False
* make inplace = True if you want overwrite the result

```python
data = pd.DataFrame()
data['Names']=['Rahil','Sahil','Kamil','Shagil']
data['Age']=[18,17,16,24]
data['city']=['Lko','Kanpur','Barabanki','Gonda']
data['marks'] = [90,96,50,44]
data
```

```python
data.drop(index = 1, inplace=True)
```

```python
data
```

**rename the column**

```python
data = pd.DataFrame()
data['Names']=['Rahil','Sahil','Kamil','Shagil']
data['Age']=[18,17,16,24]
data['city']=['Lko','Kanpur','Barabanki','Gonda']
data['marks'] = [90,96,50,44]
data
```

```python
data.rename({'city':'City'},axis=1)
```

```python
data
```

```python
data.rename(mapper={'Names':'names','Age':'age'},axis=1,inplace=True)
```
