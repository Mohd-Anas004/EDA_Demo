```python
plt.figure(figsize=(30,12))
for i in range(1,len(cat)):
   keys = visadf[cat[i]].value_counts().keys()
   value = visadf[cat[i]].value_counts().values
   plt.subplot(3,3,i)
   plt.bar(keys,value)
   
   plt.xlabel('class name')
   plt.ylabel('frequency count')
   plt.title(f'{cat[i]}_barchart')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))
for i in range(1,len(cat)-1):
    keys = visadf[cat[i]].value_counts().keys()
    values = visadf[cat[i]].value_counts().values
    plt.subplot(3,3,i+1)
    plt.pie(x = values,labels=keys,radius=1)
```

```python
plt.figure(figsize=(20,20))
for i,j in enumerate(cat[1:]):
    keys = visadf[j].value_counts().keys()
    value = visadf[j].value_counts().values
    plt.xlabel('Class Name')
    plt.ylabel('Frequency Name')
    plt.title(f'{j}_barchart')
    plt.subplot(3,3,i+1)
    plt.bar(x=keys,height=value)
plt.tight_layout()
plt.show()
```

```python
import os 
folder_name = 'Bar_chart'
folder = os.makedirs(folder_name,exist_ok=True)
directory = os.getcwd()
new_dir = os.path.join(directory,folder_name)
plt.figure(figsize=(30,8))
for i in cat[1:]:
    label = visadf[i].value_counts().keys()
    values = visadf[i].value_counts().values
    plt.bar(x = label,height=values)
    plt.xlabel('Class')
    plt.ylabel('Frequency')
    plt.title(f'{i}_barchart')
    filename = f'{i}_barchart'
    path = os.path.join(new_dir,filename)
    plt.savefig(path)
```

```python
import os 
folder_name = 'Pie_chart'
folder = os.makedirs(folder_name,exist_ok=True)
directory = os.getcwd()
new_dir = os.path.join(directory,folder_name)
plt.figure(figsize=(12,8))
for i in cat[1:]:
    label = visadf[i].value_counts().keys()
    values = visadf[i].value_counts().values
    plt.pie(x = values,labels=label,autopct='%1.2f%%')
    plt.title(f'{i}_piechart')
    filename = f'{i}_piechart'
    path = os.path.join(new_dir,filename)
    plt.savefig(path)
```
