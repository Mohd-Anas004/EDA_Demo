**Encoding**
- encoding means changing cat data to num data
- ML doesnt understand english char so
- it is very important to convert the categorical labels to numerical data
- for example we have two columns
    - Male
    - Female
- Male represent 1
- Female represent 0
- based on alphabetical orders

**One Hot Encoder**

- First how many column are there in a column
- That many new columns will created extra
- far ex case status has two labels
    - Certified
    - Denied
- So it will create two extra new column
    - case_status_Certified
    - case_status_Denied
- At a time one column value becomes 1 other columns value becomes 0
- This means a **One Hot Encoder**   

#for ex
|case_status|case_status_Certified|case_status_Denied
|-|-|-|
Certified|1|0|
Denied|0|1
